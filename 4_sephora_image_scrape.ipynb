```python
urls = list(df_products_detailed.image_url.values)
filenames = [filename+".png" for filename in df_products_detailed.id.values]
downloaded_images = [file for file in os.listdir(image_folder) if file.endswith(".png")]
```

```python
new_products={}

for filename, url in zip(filenames, urls):
    if filename not in downloaded_images:
        new_products[filename]=url

len(new_products)
```

```python
for filename,url in tqdm(new_products.items()):
    scrape.get_image(url, filename, image_folder)
```
